```python
from fastai.vision.all import *

# Download and use the fastai MNIST_SAMPLE dataset, which is organized as images in folders
path = untar_data(URLs.MNIST)

mnist_block = DataBlock(
    blocks=(ImageBlock(cls=PILImageBW), CategoryBlock),
    get_items=get_image_files,
    splitter=RandomSplitter(valid_pct=0.2),
    get_y=parent_label,
    batch_tfms=[*aug_transforms(), Normalize.from_stats(*mnist_stats)]
)

dls = mnist_block.dataloaders(path)
```

```python
dls.show_batch()
```

```python
learn = cnn_learner(dls, xresnet18, n_in=1, metrics=accuracy)
learn.fine_tune(5)
```

```python
learn.show_results()
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```

```python
learn.lr_find()
```

```python
learn.recorder.plot_lr_find()
```

```python
learn.fine_tune(5)
learn.unfreeze()
learn.fit_one_cycle(5, slice(1e-6, 1e-3))
```

```python
if len(dls.train) == 0 or len(dls.valid) == 0:
    raise ValueError("DataLoaders are empty. Check your dataset path and DataBlock setup.")

dls.show_batch(max_n=4)

learn.lr_find()
learn.recorder.plot_lr_find()
```

```python
import gradio as gr
from fastai.vision.all import load_learner, PILImage

# Load the trained model
learn = load_learner('mnist_model.pkl')

def predict(img):
    img = PILImage.create(img)
    pred_class, pred_idx, probs = learn.predict(img)

    # Only accept predictions of "3" or "7"
    if pred_class in ["3", "7"]:
        return f"✔️ Correct: {pred_class}, Confidence: {probs[pred_idx]:.4f}"
    else:
        return f"❌ Error: Model predicted {pred_class}, but only 3 or 7 are valid!"

interface = gr.Interface(
    fn=predict,
    inputs=gr.Image(image_mode='L'),
    outputs="text",
    title="MNIST Classifier",
    description="Upload an image of a digit. Only 3 and 7 are valid.",
)

interface.launch()
```
